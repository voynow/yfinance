```python
import matplotlib.pyplot as plt
import pandas as pd
import utils

plt.style.use('ggplot')
```

```python
symbols = ["^VIX", "SPY"]
data = utils.yf_download(symbols)
data.to_csv('SPY_VIX_1h.csv')
```

```python
fig, ax1 = plt.subplots(figsize=(14, 4))
ax2 = ax1.twinx()

ax1.plot(data['SPY_Close'], c="#6666cc", linewidth=2, alpha=.85)
ax2.plot(data['VIX_Close'], c="#cc6666", linewidth=2, alpha=.85)

ax1.set_xlabel("Date", fontsize=14)
ax1.set_ylabel("SPY", color="#5555cc", fontsize=16)
ax2.set_ylabel("VIX", color="#cc5555", fontsize=16)

plt.show()
```
